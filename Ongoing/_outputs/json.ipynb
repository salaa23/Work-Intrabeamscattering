```python
import json
import matplotlib.pyplot as plt
import numpy as np
import ipympl
```

```python
%matplotlib ipympl

file_name = "output_250411_091713.json"
with open("output_250411_114658.json") as f:
    outputs = json.load(f)

print(outputs.keys())
emitx = outputs["epsx"]
print(type(emitx["value"]))
emit_x = np.array(emitx['value'])

plt.cla()
plt.plot(emit_x*1e12)
plt.ylabel("emittance (m)")
plt.xlabel("turns")
plt.title("emittance plot from xsuite" )
plt.show()
```
